# Import Librarise:

```python
# importing required Libraries 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Data Points:

```python
# creating data random points 
x = np.round(np.random.uniform(5, 25, 10))
y = x*2
```

```python
print(y, x)
```

# Predict X values:

```python
# function to predict x values

def predict_y(x, m=1, c=0):
    return m*x+c
```

```python
y_pred = predict_y(x)
```

```python
y_pred
```

# Actual vs Predicted values:

```python
# comparing actual vs predicted by ploting graph

plt.figure(figsize=(10, 5))
plt.scatter(x, y, label="actual")
plt.plot(x, y_pred, label="predicted")
plt.xlabel('X values')
plt.ylabel('Y values')
plt.legend()
plt.savefig('lr.jpg')
plt.show()
```

# Cost Funtion:

```python
# defining cost function

def cost_function(y_pred, y):
    return (np.sum(np.square(y_pred-y)))/(2*len(y))
```

```python
cost_function(y_pred, y)
```

```python
# another way find cost fuction

sum = 0
for y1, y2 in zip(y_pred, y):
    sum+=(y1-y2)**2
sum/(2*len(y))
```

# Cost Function vs m

```python
# ploting graph for diffrent values of m vs cost function

plt.figure(figsize=(10,5))

M = [0, 0.5, 1, 1.5, 2, 2.5, 3]
for m in M:
    error = cost_function(predict_y(x, m), y)
    
    print(f'for m = {m} error is {error}')
    plt.scatter(m, error)
    
plt.ylabel("cost function")
plt.xlabel("m")
plt.title("Error vs m")
plt.savefig('costfunc.jpg')
plt.show()
```

# Predicted vs m

```python
# ploting graph for diffrent values of m vs predicted y values


plt.figure(figsize=(10,5))

def plot_m(m, x):
    plt.plot(x, predict_y(x, m), label=f'm={m}')


for m in M:
    plot_m(m, x)
    
plt.title('Diffrent values of M')
plt.legend()
plt.savefig('diff_M.jpg')
plt.show()
```

# Implementation:

```python
# implementing the linear regression

from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x.reshape(-1,1), y)
pred = lr.predict(x.reshape(-1,1))

print(y, pred, sep='\n')
    
```
